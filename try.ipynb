```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("try.csv") 
print(df.head(3))
```

```python
df.rename(columns={'Image Index':'img_index','Finding Labels':'labels','Follow-up #':'follow_up','Patient ID':'pa_id',
                   'Patient Age':'pa_age','Patient Gender':'pa_gender','View Position':'view_position'}, inplace=True)

Age_below_117 = df['pa_age'] < 117 
df[Age_below_117]
tidy_df=df[Age_below_117]
tidy_df.head()
```

```python
print(tidy_df['pa_age'].max() , tidy_df['pa_age'].min())
```

```python
df_agegroup=tidy_df.groupby('pa_age').count().reset_index()
df_agegroup
```

```python
df_new= df_agegroup.drop(columns=['labels','follow_up','pa_id','pa_gender','view_position','OriginalImage[Width','Height]','OriginalImagePixelSpacing[x','y]'])
df_new.head()
```

```python
bins = [55, 65, 75, np.inf]
labels=['55-65','66-75','more than 200']
tidy_df=tidy_df.groupby(pd.cut(tidy_df['pa_age'], bins=bins, labels=labels)).size().reset_index(name='count')
print (tidy_df)
        
#https://stackoverflow.com/questions/48280733/how-to-group-by-list-ranges-of-value-in-python-pandas
```
